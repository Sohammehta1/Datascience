```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_csv('iris.csv')
```

```python
df.columns =['sepal_length', 'sepal_width','petal_length','petal_width','flower_class']
df.loc[len(df.index)]=[5.1,3.5,1.4,0.2,'Iris-setosa'] #Since 1st row was replaced by column names.
```

```python
df.isna().sum()
```

```python
for cols in df.columns:
    print(f" {cols}: {type(df[cols][4])}")
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x= 'sepal_length', hue='flower_class')
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x= 'sepal_width', hue='flower_class')
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x= 'petal_length', hue='flower_class')
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x= 'petal_width', hue='flower_class')
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x= 'flower_class', hue='flower_class')
```

```python
sns.boxplot(data=df, y = 'sepal_length',hue='flower_class')
```

```python
sns.boxplot( x = df['petal_width'], y = 'count',hue='flower_class')
sns.boxplot(data=df, x = 'sepal_width',hue='flower_class')
```

```python
sns.boxplot(data=df, x = 'petal_length',hue='flower_class')
```

```python
sns.countplot(data=df, x='petal_width', hue='flower_class')
plt.title('Count of Flower Class by Petal Width')
plt.xlabel('Petal Width')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
