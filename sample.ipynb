```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Anti_dataset.csv')

# Function for equal-width partitioning
def equal_width_partitioning(df, col_name, num_bins):
    width = (df[col_name].max() - df[col_name].min()) / num_bins
    equal_width = []
    for i in range(num_bins):
        group = df[(df[col_name] >= i * width) & (df[col_name] < (i + 1) * width)][col_name].tolist()
        equal_width.append(group)
    # For the last bin, include values up to the maximum
    equal_width.append(df[df[col_name] >= num_bins * width][col_name].tolist())
    return equal_width

def equal_depth_partitioning(df, col_name, num_bins):
    sorted_col = sorted(df[col_name])
    bin_size = len(df) // num_bins
    bins = [sorted_col[i * bin_size: (i + 1) * bin_size] for i in range(num_bins)]
    # Add remaining elements to the last bin
    bins[-1] += sorted_col[num_bins * bin_size:]
    
    equal_depth = []
    for i, bin_values in enumerate(bins):
        equal_depth.append(bin_values)
    return equal_depth

num_bins = 5

equi_width = equal_width_partitioning(df, 'V1', num_bins)
equi_depth = equal_depth_partitioning(df,'V1',num_bins)
# equi_depth = equal_depth_partitioning(df, 'V1', num_bins)

# Display the updated dataframe
print("Equi width ")
for i in range(num_bins):
    print(f"group{i+1} : {equi_width[i]}")

print("Equi depth")
for i in range(len(equi_depth)):
    print(f"group{i+1} : {equi_depth[i]}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from scipy.stats import pearsonr



for i, (width_group, depth_group) in enumerate(zip(equi_width, equi_depth), 1):
    print(f"\nGroup {i} Analysis:")
    
    # a. Compute mean, Q1, median, Q3, and standard deviation of the T-virus measures
    t_virus_stats = pd.DataFrame({'V1': depth_group}).describe().T
    print("T-virus Statistics:")
    print(t_virus_stats)

    # b. Draw boxplots of the T-virus measures
    plt.figure(figsize=(8, 5))
    plt.boxplot([width_group, depth_group], labels=['Equal Width', 'Equal Depth'])
    plt.title(f"Boxplot of T-virus Measures - Group {i}")
    plt.ylabel("T-virus")
    plt.show()

    if len(width_group) > 0 and len(depth_group) > 0:
    # c. Normalize Anti-A and Anti-B amounts based on min-max normalization
        scaler = MinMaxScaler()
        normalized_width = scaler.fit_transform(pd.DataFrame({'ANTI_A': width_group, 'ANTI_B': width_group}))
        normalized_depth = scaler.fit_transform(pd.DataFrame({'ANTI_A': depth_group, 'ANTI_B': depth_group}))
        
        print("\nNormalized Anti-A and Anti-B Values:")
        print(f"Equal Width Group:\n{normalized_width}")
        print(f"Equal Depth Group:\n{normalized_depth}")

        # d. Draw scatter plots of normalized Anti-A and normalized Anti-B
        plt.figure(figsize=(8, 5))
        plt.scatter(normalized_width[:, 0], normalized_width[:, 1], label='Equal Width', alpha=0.7)
        plt.scatter(normalized_depth[:, 0], normalized_depth[:, 1], label='Equal Depth', alpha=0.7)
        plt.title(f"Scatter Plot of Normalized Anti-A and Anti-B - Group {i}")
        plt.xlabel("Normalized Anti-A")
        plt.ylabel("Normalized Anti-B")
        
        # e. Calculate Pearson correlation coefficient between normalized Anti-A and normalized Anti-B
        corr_width, _ = pearsonr(normalized_width[:, 0], normalized_width[:, 1])
        corr_depth, _ = pearsonr(normalized_depth[:, 0], normalized_depth[:, 1])
        print("\nPearson Correlation Coefficients:")
        print(f"Equal Width Group: {corr_width}")
        print(f"Equal Depth Group: {corr_depth}")

        # f. Conclusion
        if corr_width > corr_depth:
            conclusion = f"For Group {i}, using equal-width partitioning is more effective in fighting T-virus."
        else:
            conclusion = f"For Group {i}, using equal-depth partitioning is more effective in fighting T-virus."
        print(conclusion)

plt.legend()
plt.show()


```
