```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.xls')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Gender'] =  df.Gender.astype('category')
```

```python
df['Gender'] = df['Gender'].cat.codes
```

```python
print(df)
```

```python
X = df.drop(columns=['Purchased','User ID','Gender'])
Y = df['Purchased']
```

```python
print(X)
```

```python
fig,axes = plt.subplots(1,3,figsize = (14,4))
sns.countplot(data = df, x = 'Purchased',hue='Gender',ax = axes[0])
sns.scatterplot(data =df,x = 'Age',y='Purchased',hue='Age',ax =axes[1])
sns.scatterplot(data =df,x = 'EstimatedSalary',y='Purchased',ax =axes[2])
```

```python
sns.heatmap(df.corr(),annot=True)
```

## From the above we can observe that, Purchased has good correlation with Age and slightly less with Estimated salary

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


x_train,x_test, y_train,y_test = train_test_split(X,Y, train_size=0.8, random_state=0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_pred,y_test)
print(acc)
```

```python
from sklearn.metrics import precision_score, confusion_matrix,recall_score,f1_score

cm = confusion_matrix(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print(cm)
print(precision)
print(recall)
print(f1)
```
