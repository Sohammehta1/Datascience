```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('HousingData.csv')
```

```python
df
```

```python
df.isna().sum()B: B represents the proportion of Black residents by town. It indicates the percentage of the population that identifies as Black or African American.
```

```python
for col in df.columns:
    mean_val = df[col].mean(skipna=True)
    df[col].fillna(mean_val,inplace =True)
```

```python
df.isna().sum()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data = df, x= 'B', y = 'MEDV')
```
