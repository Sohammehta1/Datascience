```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.describe()
```

```python

df['math score'].replace('?', np.nan, inplace=True)
df['math score'] = df['math score'].astype(float)
math_score  = df['math score'].clip(0.0,100.0)
sns.histplot(math_score,bins=25,kde=True)
plt.axvline(math_score.mean(), color = 'k', linestyle='dashed', linewidth=1)
plt.show()
```
