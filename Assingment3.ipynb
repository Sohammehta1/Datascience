Analysis : 1)Comparison of sepal width and sepal length -> With increase in sepal length, the sepal width of virginica and versicolor seems to be increasing, they seem to have positive correlation between the two.
2)setosa seems to be having the largest petal width, but is the smallest in all other attributes.
3)The Attributes for virginica and versicolor are all giving positive correlation.

```python

setosa = df[df['flower_class'] == 'Iris-setosa'].drop('flower_class',axis=1)
virginica = df[df['flower_class'] == 'Iris-virginica'].drop('flower_class',axis=1)
versicolor = df[df['flower_class'] == 'Iris-versicolor'].drop('flower_class',axis=1)

setosa_filtered = pd.DataFrame()
virginica_filtered = pd.DataFrame()
versicolor_filtered = pd.DataFrame()

for col in setosa.columns:
    q1 = np.percentile(setosa[col], 25)
    q3 = np.percentile(setosa[col], 75)
    iqr = q3 - q1
    setosa_filtered = setosa[(setosa[col] >= (q1 - 1.5 * iqr)) & (setosa[col] <= (q3 + 1.5 * iqr))]

for col in virginica.columns:
    q1 = np.percentile(virginica[col], 25)
    q3 = np.percentile(virginica[col], 75)
    iqr = q3 - q1
    virginica_filtered = virginica[(virginica[col] >= (q1 - 1.5 * iqr)) & (virginica[col] <= (q3 + 1.5 * iqr))]

for col in versicolor.columns:
    q1 = np.percentile(versicolor[col], 25)
    q3 = np.percentile(versicolor[col], 75)
    iqr = q3 - q1
    versicolor_filtered = versicolor[(versicolor[col] >= (q1 - 1.5 * iqr)) & (versicolor[col] <= (q3 + 1.5 * iqr))]
 
```

```python
print(setosa.describe())
print(setosa_filtered.describe())
```

```python


fig2, axes2 = plt.subplots(1, 4, figsize=(15, 5))

setosa_filtered = np.array(setosa_filtered)
virginica_filtered = np.array(virginica_filtered)
versicolor_filtered = np.array(versicolor)

# First subplot
axes2[0].scatter(setosa_filtered[:, 0], setosa_filtered[:, 1], color='blue', label='setosa')
axes2[0].scatter(virginica_filtered[:, 0], virginica_filtered[:, 1], color='yellow', label='virginica')
axes2[0].scatter(versicolor_filtered[:, 0], versicolor_filtered[:, 1], color='red', label='versicolor')
axes2[0].set_ylabel('sepal_width')
axes2[0].set_xlabel('sepal_length')
axes2[0].legend()

# Second subplot
axes2[1].scatter(setosa_filtered[:, 2], setosa_filtered[:, 3], color='blue', label='setosa')
axes2[1].scatter(virginica_filtered[:, 2], virginica_filtered[:, 3], color='yellow', label='virginica')
axes2[1].scatter(versicolor_filtered[:, 2], versicolor_filtered[:, 3], color='red', label='versicolor')
axes2[1].set_ylabel('petal_width')
axes2[1].set_xlabel('petal_length')
axes2[1].legend()

# Third subplotfiltered_setosa
axes2[2].scatter(setosa_filtered[:, 0],setosa_filtered[:, 2], color='blue', label='setosa')
axes2[2].scatter(virginica_filtered[:, 0], virginica_filtered[:, 2], color='yellow', label='virginica')
axes2[2].scatter(versicolor_filtered[:, 0], versicolor_filtered[:, 2], color='red', label='versicolor')
axes2[2].set_xlabel('sepal_length')
axes2[2].set_ylabel('petal_length')  # Corrected label
axes2[2].legend()

# Fourth subplot
axes2[3].scatter(setosa_filtered[:, 1], setosa_filtered[:, 3], color='blue', label='setosa')
axes2[3].scatter(virginica_filtered[:, 1], virginica_filtered[:, 3], color='yellow', label='virginica')
axes2[3].scatter(versicolor_filtered[:, 1], versicolor_filtered[:, 3], color='red', label='versicolor')
axes2[3].set_xlabel('sepal_width')
axes2[3].set_ylabel('petal_width')
axes2[3].legend()
```
